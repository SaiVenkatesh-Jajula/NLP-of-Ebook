### To get accurate result - regex entries!

```python
import re
pattern=re.compile("Chapter [0-9]")
```

```python
# The above is just a pattern object
re.findall(pattern,book)
```

```python
# To get multiple digits 
pattern=re.compile("Chapter [0-9]+")
print(pattern)
```

```python
chapters=re.findall(pattern, book)
chapters
```

```python
# Now count the no. of chapters
len(chapters)
```

# What are the most used words in the book?

```python
# getting all the words in the book
# pattern = re.compile("[a-zA-Z][^ .]*")
pattern = re.compile("[a-zA-Z]+")
findings=re.findall(pattern,book.lower())
print(len(findings))
print(findings[:5])
```

```python
# couting the occurance of each word
dictionary={}
for item in findings:
    if item in dictionary.keys():
        dictionary[item] = dictionary[item]+1
    else:
        dictionary[item] = 1
```

```python
newlist = [(item,key) for key,item in dictionary.items()]
print(sorted(newlist, reverse=True))
```

## Function that finds the occurance of any word.

```python
# couting the occurance of each word
dictionary={}
for item in findings:
    if item in dictionary.keys():
        dictionary[item] = dictionary[item]+1
    else:
        dictionary[item] = 1
def find(word):
    try:
        return dictionary[word]
    except KeyError:
        return f"The book does not contain the word : {word}"
```

```python
#call the function
find('love')
```

```python
find('hate')
```

```python
pattern = re.compile("[^\n]*[A-Z][a-z]*[\n]{2}")
titles = re.findall(pattern, book)
titles = [i.strip('\n\n') for i in titles]
titles
```
